```python
with open(os.path.join('..', 'data', 'train.csv'), 'rb') as csvfile:
    training_data = np.loadtxt(csvfile, delimiter=',')
```

```python
row = 5
col = 5

indexes = np.random.choice(len(training_data[:,0]), row * col, replace = False)

fig = plt.figure(figsize=(10,10))

for i, img in enumerate(training_data[indexes,:]):    
    fig.add_subplot(row, col, i+1)
    plot = plt.imshow(img[9:].reshape([33,13]), cmap='gray')
    plt.axis('off')

print('Truth:')
for img in training_data[indexes,:]:
    print(int(img[1]))

print('\nBook:')
for img in training_data[indexes,:]:
    print(int(img[2]))

print('\nId:')
for img in training_data[indexes,:]:
    print(int(img[0]))
```
